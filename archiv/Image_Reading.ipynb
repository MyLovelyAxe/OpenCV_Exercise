```python
import cv2
import matplotlib.pyplot as plt
```

### Image reading & showing

```python
''' show with matplotlib '''
# to show image with matplotlib, it is necessary to convert color scale from bgr (opencv) to rgb (matplotlib)
im = cv2.imread('redhood.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
print(im.shape)
fig = plt.figure()
ax = plt.subplot()
ax.imshow(im,cmap='Reds')
plt.show()
```

```python
''' show with cv2.imshow() '''
im = cv2.imread('redhood.jpg')
print(im.shape)
cv2.imshow('hahaha',im) # arguments: window_name, image
cv2.waitKey(0) # 0: window remains open until any input from keyboard interferes
cv2.destroyWindow('hahaha') # in order to close window successfully in jupyter notebook, this line is necessary for closing corresponding window
# destroyAllWindows() # or just close all windows
```

### Video reading & showing

```python
# choose a specific video with its name
cap = cv2.VideoCapture('whatever.mp4') 
while True:
    success, img = cap.read() # read frames one by one, return img and whether it is successful
    cv2.imshow('Video', img) # show the current frame
    if cv2.waitKey(1) & 0xFF == ord('q'): # wait for specific keyboard 'q' to close the window
        break
```

```python
cap = cv2.VideoCapture(0) # choose webcam oject with 0, if only one webcam is available
cap.set(3,640) # set height, whose id is 3 in setting
cap.set(4,480) # set width, whose id is 4 in setting
cap.set(10,100) # set brightness, whose id is 10 in setting
while True:
    success, img = cap.read() # read frames one by one, return img and whether it is successful
    cv2.imshow('Video', img) # show the current frame
    if cv2.waitKey(1) & 0xFF == ord('q'): # wait for specific keyboard 'q' to close the window
        break
```
